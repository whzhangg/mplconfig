## A simple plot using Matplotlib
### Introduction
The following example illustrates simple case of plotting a function. To plot an figure, we first import `matplotlib` packages, as well as `numpy`. We create `x` data using numpy's [linear space method](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html) and `y` data as a $\sin(x)$ function. 

In Matplotlib, different elements are separate objects. For example, in this simple plot, we have two objects: a figure `fig` and an axis `axs`. The figure object is the "canvas" on which we can plot data. But we do not plot data directly on figures. Instead, we plot data on an axis object. An axis object include data, frame, tics and legends. To make distinction between figure and axis: a figure can contain multiple axis showing different data, as I will later show. 

### Figures
A matplotlib figure is created by the `plt.figure()` method. When creating figure, we can pass some optional arguments, for example, the figure size. The [matplotlib page documentation on figure](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) shows more detail on these arguments. One useful optional arguments is `figsize`. For example,
```python
fig = plt.figure(figsize = (3,1.8))
```
will create a figure with width of 3in and height of 1.8in. The values in `figsize` argument are given in a tuple of `(w,h)`, however, the values of width and height depend on output format. For example, for pdf output, the unit of figure size is in inch, but for png or jpg output, the unit of figure size will be in pixels.

For publication, journals always have requirement for the figure size depending on if the figure is single or double column. For example, a single column figure for ACS should be about 3 inch in width, as given in the [author guideline](https://publish.acs.org/publish/author_guidelines?coden=cmatex#preparing_graphics). It is suggested to export figures in vector format, such as `.pdf`. They can be converted to `.png` files with desired pixel density (for example, 300ppi) if needed.

### Axis
We need a figure to create an axis, this is given by `axs = fig.subplots()` documented [here](https://matplotlib.org/stable/api/figure_api.html?highlight=subplots#matplotlib.figure.Figure.subplots). Note that this function can also take more arguments, which we will describe later. 

Next thing we did is to set some parameters to the axis. We set the x and y limits, as well as the labels for x and y. Notice that Matplotlib can use the syntax of *latex* to enter inline equations. This is the way to enter greek symbols or mathematical equations in the plot. This ability is not restricted to axis labels. We can use latex syntax in labels or text boxs in the figure. Also notice that when using latex syntax, I added a `r` in front of the string: `r""`. This is the use of [raw string](https://note.nkmk.me/en/python-raw-string-escape/). In latex syntax, `\` character are frequently used and sometimes this will confuse python. Therefore, it is suggested to use raw string when latex syntax is used.

### Plotting data
Finally, we plot the data using `axs.plot()` function. Two lines are plotted, one line is the actual data and the second line is the base line showing $y=0$. Apart from the x and y data values, we supplied many other parameters to the plot function. Typically, three arguments are supplied as minimum: x, y and line types. Line types can be specified using symbols (fmt) like `'*-'`, `'o--'` and so on. The first character determines the shape of the marker. For example, `'*'` stands for stars, `'o'` stands for cycles, `'-'` stands for solid lines and `'--'` stand for dashed line. Complete description of the plot function and the format symbol can be found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html). If you want to plot lines only, gave only the format symbol for lines (`'-'`), if you want to plot points only, use only the format symbol for points (`'o'`).

To control color, we specify `color="colorname"`, Matplotlib defined many color name [here](https://matplotlib.org/stable/gallery/color/named_colors.html). If color option is not specified, Matplotlib will use predefined colors (blue, orange etc.). `alpha` parameter specifies the transparence of the plotted lines. A smaller value means more transparent value. Finally, we can assign label to each plotted lines using the keyword `label`, which take a string argument. However, labels are not necessary.

### Legends
We give labels to each plotted lines, but giving the label does not mean that they will be displayed. For that we need to ask for a legend box. This is added by `axs.legend()` method. It is documented [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html). This method can take some arguments, the most important one being the `frameon` argument, which disables the default boarder. It also accept a `loc` argument which allow you to manually specify a location. Location is specified using a combination of keywords. To specify the vertical location, use "upper", "center" or "lower". For horizontal locations, use "left", "center", "right". The default value for location is "best", which ask Matplotlib determine the location depending on available space. 

### Saving figures
To save figures to a file, we call the `savefig()` function from the figure objects. A filename need to be specified as arguments. Matplotlib will infer the file type and export the figures to the given file type. However, one point to be cautious is that setting of the figure will depend on the choice of output file format, especially for various size parameter. The most important one is the figure size parameter, as described before. For other parameter, it's better to adjust after you see the result of the output.

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.01, 5, 100)
y = np.sin(1.0 / x)

fig = plt.figure()
axs = fig.subplots()
axs.set_xlim(0, 5)
axs.set_ylim(-1.1, 1.1)
axs.set_xlabel("x value")
axs.set_ylabel(r"y value ($\sin \frac{1}{x}$)")
axs.plot(x, y, "*-", color = "darkgreen", label = "f(x)")
axs.plot([0, 10], [0, 0], '--', color = "gray", alpha = 0.5)
axs.legend(frameon = False, loc = "upper right")
fig.savefig("simple_figure.pdf")
````
