## Multiplots in Matplotlib
### Concepts
As described above, in Matplotlib, figure object contains the axis object. To make a multiplot, we essentially is asking the figure to include multiple axes and we can create different plots at different axis. Therefore, the most important step in creating a multiplot is just to create different axes. After axes are created, the plotting process is just the same as we already described and all we need to do is to go over all axis.

### Using `figure.subplots()`
The simplest way is to use `figure.subplots()` method ([subplots](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.subplots)), which return axes. We have already used this method previously but in that case only a single axis is returned by default. To return multiple axes, we use:
```python
axes = fig.subplots(nrows, ncols)
```
for a multiplot figure with nrows and ncols. Notice that in this case, the return is an array (multidimension list, as a numpy array) of axes. To plot data to each axis, we should specify index: `axes[0,0].plot(x, y)`. 

### Using `figure.add_axes()`
A slightly more complicated way to make multiple figure is to use [add_axes](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_axes) method, where we add axis one by one with specified size ratio. The dimension of the axis area is passed as a tuple of (left, bottom, width, height) with values in terms of the ratio compared to figure width and height. This method has several advantages: firstly, it in general give more flexibility to multiplot, for example, when the boarder or plot separation is too large or small. It is also more straightforward to use compare to `subplots()` where Matplotlib determines for you the suitable axes position, which could be not very ideal. Finally, when you specify different figure size and ratio, it is convenient to use `add_axes()` to make plots look consistent. For example, a plot in a single-column figure can be adjusted to have the same size as a plot in a double-column multiplot figure. This makes the plots look consistent across papers.

Just to give an example on how to use `add_axes()` method. Consider if you want to add an axis to the left size of the figure, use:
```python
axis_left = fig.add_axes([0.12, 0.12, 0.4, 0.8])
```
so that the plot will start at 0.12 (fraction of the figure size) from the left and bottom of the figure edge, and will have the size $0.4 \times 0.8$ which span most part of the left size. One thing to note is that often axes is added with more space in the bottom and left to contain the tics and labels, but less space is needed on the right side and top side. 

````python
# Example using `figure.subplots()`
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y1 = np.exp(x)
y2 = x**3

figure = plt.figure(figsize=(10,4))
axes = figure.subplots(1, 2)

axes[0].plot(x, y2, label = "$x^3$")
axes[0].plot(x, np.zeros_like(x), color = "gray")
axes[0].set_xlabel("x variable")
axes[0].set_ylabel("y variable")
axes[0].legend()

axes[1].plot(x, y1, label = "$\exp(x)$")
axes[1].plot(x, np.zeros_like(x), color = "gray")
axes[1].set_xlabel("x variable")
axes[1].set_ylabel("y variable")
axes[1].legend()
````

````python
# Example 2, using add_axes method
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y1 = np.exp(x)
y2 = x**3

figure = plt.figure(figsize=(10,4))
axes1 = figure.add_axes([0.1, 0.15, 0.37, 0.8])
axes2 = figure.add_axes([0.55, 0.15, 0.37, 0.8])

axes1.plot(x, y2, label = "$x^3$")
axes1.plot(x, np.zeros_like(x), color = "gray")
axes1.set_xlabel("x variable")
axes1.set_ylabel("y variable")
axes1.legend()

axes2.plot(x, y1, label = "$\exp(x)$")
axes2.plot(x, np.zeros_like(x), color = "gray")
axes2.set_xlabel("x variable")
axes2.set_ylabel("y variable")
axes2.legend()
````
